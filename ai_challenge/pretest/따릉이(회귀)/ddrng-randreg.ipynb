# Load data

```python
train = pd.read_csv('/kaggle/input/sejongai-challenge-pretest-2/2020.AI.bike-train.csv')
test = pd.read_csv('/kaggle/input/sejongai-challenge-pretest-2/2020.AI.bike-test.csv')
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
lr_reg = LinearRegression()
ridge_reg=Ridge(alpha=10)
lasso_reg=Lasso(alpha=0.01)
RFR = RandomForestRegressor(n_estimators=500, max_depth=3,random_state=1123)

models = [lr_reg, ridge_reg, lasso_reg, RFR]
```

```python
y = train.pop('count')
```

```python
X_train, X_test, y_train,y_test = train_test_split(train, y, test_size=0.2, random_state=1123)
```

```python
def model_predict(md):
    md.fit(X_train,y_train)
    pred = md.predict(X_test)
    print(i, metrics.mean_squared_error(y_test,pred))
```

```python
for i in models:
    model_predict(i)
```

```python
# models = [lr_reg, ridge_reg, lasso_reg, RFR]
pred = RFR.predict(test)
```

```python
pred_noround = pred
pred_round = np.round(pred)
```

```python
idx = list(range(0,len(pred)))
```

```python
sub1 = pd.DataFrame({'id':idx,'count':pred_noround})
sub2 = pd.DataFrame({'id':idx,'count':pred_round})
```
